```python
from langchain_openai import ChatOpenAI
from langchain_community.tools import StructuredTool, DuckDuckGoSearchRun, tool
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_classic import hub
from pydantic import Field, BaseModel
```

```python
search_tool = DuckDuckGoSearchRun()

llm = ChatOpenAI(base_url="https://lightning.ai/api/v1/", api_key="API_KEY")
```

```python
search_tool.invoke("What is Date today")
```

```python
prompt = hub.pull('hwchase17/react')
```

```python
import re

@tool
def sum_tool(numbers: str) -> int:
    """Takes input like '100, 34' and returns their sum"""
    nums = re.findall(r'\d+', numbers)
    a, b = map(int, nums[:2])
    return a + b
```

```python
agent = create_react_agent(
    llm=llm,
    tools=[search_tool, sum_tool],
    prompt=prompt
)
```

```python
agent_excutor = AgentExecutor(
    agent=agent,
    tools=[search_tool, sum_tool],
    verbose=True
)
```

```python
output = agent_excutor.invoke({"input" : "What Is Capital of Uganda And Also Tell me What is Sum og 68 and 2"})
```
